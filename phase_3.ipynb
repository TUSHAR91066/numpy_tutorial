```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure: [restaurrant_id , 2021 , 2022 , 2023, 2024]
sales_data = np.array([
    [1 , 150000, 180000, 220000, 250000], # Paradise biryani
    [2 , 120000, 140000, 160000, 190000], # Beijing bites
    [3 , 200000, 230000, 260000, 300000], # Pizza hub
    [4 , 180000, 210000, 240000, 270000], # burger Point
    [5 , 160000, 185000, 205000, 230000]  #chai point 
])

print("==== Zomato sales analysis ====")
print("\n sales data shape", sales_data.shape)
print("\n sample data for 1st 3 restaurant: ", sales_data[:,1:])
```

```python
# total sales per year
print(np.sum(sales_data, axis = 0))
yeary_total = np.sum(sales_data[:, 1:], axis=1)
print(yeary_total)

# minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis = 0)
print(min_sales)

# maximum sales per year
max_sale = np.max(sales_data[:, 1:] ,axis = 1)
print(max_sale)

# Average sales per restauntant
avg_sales = np.mean(sales_data[: , 1:], axis = 1)
print(avg_sales)

# cummilative sum of the sales
cumsum = np.cumsum(sales_data[:, 1:], axis = 1)
print(cumsum)

plt.figure( figsize = (10,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales accros all the restaurant")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

```python
vector_1 = np.array([1,2,3,4,5])
vector_2 = np.array([6,7,8,9,10])

print("Vector addition: ", vector_1 + vector_2)

print("\n Multiplication vector", vector_1*vector_2)

print("\n Dot mutplication: ", np.dot(vector_1,vector_2) )

angle = np.arccos(np.dot(vector_1,vector_2)/(np.linalg.norm(vector_1) * np.linalg.norm(vector_2)))
print(angle)
```
